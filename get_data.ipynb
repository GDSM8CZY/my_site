# Get the data from the site

```python
import requests
from bs4 import BeautifulSoup

# Site where all the data is from
url = "https://dcc.dickinson.edu/latin-core-list1"

# Get html with request
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# replace "-" with word stem
def fix_PP(word, prt_of_speech):
    prt_of_speech = prt_of_speech.split(":")
    PP = word.split()
    # Only for adjective which have "-"
    if((prt_of_speech[0] == "Noun" or prt_of_speech[0] == "Adjective") and "-" in word):
        # try to replace "-" for each Principle Part with stem
        for i in list(range(len(PP))):
            if PP[0].endswith("r"):
                PP[i] = PP[i].replace("-", PP[0][0:-1])
            else:
                PP[i] = PP[i].replace("-", PP[0][0:-2])
    # connect all PP into one string and return
    return " ".join(PP)


with open("static/data/latin_core_vocabulary_list1.csv", "a", encoding="utf-8") as f:
    # write labels for cols
    f.write("word, deff, prt_of_speech, usage, frequency_rank \n")
    # find all html blocks with words
    for tr in soup.find_all('tr')[1:]:
        cols = tr.find_all('td')
        # replace ',' in deffinitions with '#'
        cols = [item.text.strip().replace(",", "#") for item in cols]
        # replace "-" with word stem
        cols[0] = fix_PP(cols[0], cols[2])
        # join all cols with ","
        out = ",".join(cols)
        # remove all the macrons
        out = out.replace("ā", "a").replace("ō", "o").replace("ū", "u").replace("ī", "i").replace("ē", "e")
        # Save latin words with all info to latin_core_vocabulary_list1.csv
        f.write(out + "\n")
```
