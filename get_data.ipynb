# Get the data from the site

```python
import requests
from bs4 import BeautifulSoup

# Site where all the data is from
url = "https://dcc.dickinson.edu/latin-core-list1"

# Get html with request
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# find all html blocks with words
with open("static/data/latin_core_vocabulary_list1.csv", "a", encoding="utf-8") as f:
    f.write("word, deff, prt_of_speech, usage, frequency_rank \n")
    for tr in soup.find_all('tr')[1:]:
        cols = tr.find_all('td')
        out = ",".join([item.text.strip().replace(",", "#") for item in cols])
        # Save latin words with all info to latin_core_vocabulary_list1.csv
        f.write(out + "\n")
```
